## 1_Splitting Brand and Model

```python
# Liste des marques disponibles
brand_list = ['Audi', 'BMW', 'Citroën', 'Mercedes-Benz', 'Peugeot', 'Renault', 'Volkswagen', 'Alfa Romeo', 'Alpina', 'Audi', 'BAIC YX', 'BMW', 'BYD', 'Chery', 'Chevrolet', 'Citroën', 'Cupra', 'Dacia', 'DFSK', 'Dodge', 'Dongfeng', 'DS', 'Fiat', 'Ford', 'Foton', 'Geely', 'GMC', 'Great Wall', 'Haval', 'Honda', 'Hyundai', 'Hyundai Trucks', 'Infiniti', 'Isuzu', 'Iveco', 'Jaguar', 'Jeep', 'KIA', 'Lada', 'Land Rover', 'Mahindra', 'Maserati', 'Mazda', 'Mercedes-Benz', 'MG', 'Mini', 'Mitsubishi', 'Nissan', 'Opel', 'Peugeot', 'Porsche', 'Renault', 'Seat', 'Skoda', 'Smart', 'Ssangyong', 'Suzuki', 'Tesla', 'Toyota', 'Volkswagen', 'Volvo', 'Wallyscar', 'ZXAUTO']

# Fonction pour séparer brand et model
def split_brand_model(row, brand_list):
    for brand in brand_list:
        if row.startswith(brand):
            # Séparer en deux parties : brand et le reste
            return brand, row[len(brand):].strip()
    return None, row  # Si aucune marque trouvée
```

```python
voitures = voitures_scraped.copy()

# Appliquer la fonction et créer deux colonnes
voitures[['Brand', 'Model']] = voitures['Modèle'].apply(
    lambda x: pd.Series(split_brand_model(x, brand_list))
)

# Supprimer la colonne 'Modèle'
voitures.drop(columns=['Modèle'], inplace=True)

# Réorganiser les colonnes pour que 'Brand' et 'Model' soient les premières
voitures = voitures[['Brand', 'Model'] + [col for col in voitures.columns if col not in ['Brand', 'Model']]]

# Afficher les premières lignes pour vérifier
voitures.head(20)
```
