```python
import requests
from bs4 import BeautifulSoup

# URL de la première page
url = "https://www.automobile.tn/fr/occasion/s=/1"

# Fonction pour extraire les informations d'une seule page
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Sélection des annonces
    articles = soup.find_all("div", class_="occasion-item-v2")

    voitures = []

    for article in articles:
        try:
            # Extraction des données principales
            model = article.find("h2").text.strip()
            kilometrage = article.find("li", class_="road").text.strip()
            annee = article.find("li", class_="year").text.strip()
            boite = article.find("li", class_="boite").text.strip()
            carburant = article.find("li", class_="fuel").text.strip()
            puissance = article.find("li", class_="horsepower").text.strip()
            prix = article.find("div", class_="price").text.strip()

            # Ajout des données dans une liste
            voitures.append({
                "Modèle": model,
                "Kilométrage": kilometrage,
                "Année": annee,
                "Boîte": boite,
                "Carburant": carburant,
                "Puissance fiscale": puissance,
                "Prix": prix,
            })
        except AttributeError:
            continue

    return voitures
```

```python
# Fonction pour scraper plusieurs pages
def scrape_all_pages(base_url, max_pages=10):
    all_voitures = []
    for page in range(1, max_pages + 1):
        print(f"Scraping page {page}...")
        page_url = base_url.replace("=/1", f"=/{page}")
        voitures = scrape_page(page_url)
        all_voitures.extend(voitures)
    return all_voitures

# Scraper toutes les pages
voitures_data = scrape_all_pages(url, max_pages=178)

# Afficher quelques résultats
for voiture in voitures_data[:5]:
    print(voiture)

# Enregistrement des résultats dans un fichier CSV
import csv

with open("scraped_voitures.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=voitures_data[0].keys())
    writer.writeheader()
    writer.writerows(voitures_data)

print("Données enregistrées dans 'voitures.csv'")
```
